```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fname = '20200101-225333.npy' # from Balboa balancer
fname = '20200102-011427.npy'
fname = ['20200102-013928.npy', '20200102-013901.npy', '20200102-014409.npy', '20200102-020345.npy', 
         '20200102-020422.npy', '20200102-020511.npy', '20200102-023107.npy', '20200102-023158.npy', 
         '20200102-023230.npy', '20200111-015402.npy']
```

```python
test = 9
data = np.load(fname[test])
data.shape
```

```python
delta_t = np.diff(data[:,0]).mean()
std_t   = np.diff(data[:,0]).std()
txt = 'delta_t {:5.3f} ms +/- {:5.3f} ms'.format(1000*delta_t, 1000*std_t)
print(txt)
```

```python
tt, mX, mY, eX, eY, gY = np.hsplit(data,data.shape[1])
```

```python
gY.mean(), gY.std(), mX.mean(), mX.std()
```

```python
plt.figure(figsize=(15,4))
plt.plot(tt, mX, 'r.-', label='mX')
#plt.plot(tt, mY, label='mY')
#plt.plot(tt, eX, label='eX')
#plt.plot(tt, eY, label='eY')
plt.plot(tt, gY, 'g.-', label='gY')
#plt.xlim(2,10)
plt.grid()
plt.title('Data from {}'.format(fname[test]))
plt.legend()
plt.show()
```
