```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fname = '20200101-225333.npy' # from Balboa balancer
fname = '20200102-011518.npy' # times are corrupted
fname = '20200102-011427.npy'
```

```python
data = np.load(fname)
data.shape
```

```python
delta_t = np.diff(data[:,0]).mean()
std_t   = np.diff(data[:,0]).std()
txt = 'delta_t {:5.3f} ms +/- {:5.3f} ms'.format(1000*delta_t, 1000*std_t)
print(txt)
```

```python
tt, mX, mY, eX, eY, gY = np.hsplit(data,data.shape[1])
```

```python
plt.figure(figsize=(15,4))
plt.plot(tt, mX, label='mX')
#plt.plot(tt, mY, label='mY')
#plt.plot(tt, eX, label='eX')
#plt.plot(tt, eY, label='eY')
plt.plot(tt, gY, label='gY')
plt.grid()
plt.legend()
```
